```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def testf(y,t):
    return [2*t, 2, 0]
tv = np.linspace(0,10,101)
y0 = [0,0, 2]
output = odeint(testf, y0, tv)
print(output)

odf = pd.DataFrame(output, columns = ['Position', 'Velocity', 'Acceleration'])
odf['idx'] = tv
odf = odf.set_index('idx')
odf.plot()
```
